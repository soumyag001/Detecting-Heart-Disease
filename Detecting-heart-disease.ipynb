# Feature

```python
#Group peoples based on age
data['age_group'] = pd.cut(data['age'], bins=[0, 30, 50, 70, 100], labels=['young', 'middle-aged', 'senior', 'elderly'])
```

```python
#Cholesterol-to-Max Heart Rate Ratio
data['chol_ratio'] = data['chol'] / data['thalach']
```

```python
#Cholesterol-to-Max Heart Rate Ratio
data['oldpeak_squared'] = data['oldpeak'] ** 2
```

```python
#Combine sex and age_group to see if certain age groups of men or women are more at risk
data['sex_age_group'] = data['sex'].astype(str) + '_' + data['age_group'].astype(str)
```

```python
#Combine thalach and exang to create a feature reflecting heart response during exercise
data['exercise_stress'] = data['thalach'] * (1 - data['exang'])
```

```python
#ST Slope Categories
data['slope_recoded'] = data['slope'].replace({1: 'up', 2: 'flat', 3: 'down'})
```

# Data Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#Binary encoding
data[["female","male"]]=OneHotEncoder().fit_transform(data[["sex"]]).toarray()
```

```python
data[["sex","female","male"]]
```

```python
print(data[['cp', 'thal']].head())
```

```python

# Apply one-hot encoding to 'cp' and 'thal' columns without manually specifying prefixes
encoded_cp_thal = pd.get_dummies(data[['cp', 'thal']])

# Concatenate one-hot encoded columns with the original DataFrame
data = pd.concat([data, encoded_cp_thal], axis=1)

# Drop the original categorical columns to avoid redundancy
data.drop(columns=['cp', 'thal'], inplace=True)

# Check the updated DataFrame
print(data.head())
```

```python
data
```

```python
#find catagorical columns for encoding
categorical_columns = data.select_dtypes(include=['object', 'category']).columns
print("Categorical Columns:", categorical_columns)
```

```python
#When initializing your OrdinalEncoder or OneHotEncoder, explicitly define all possible categories. This ensures the encoder is aware of every category it might encounter
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(categories=[['young', 'middle-aged', 'senior', 'elderly']])
data['age_group'] = encoder.fit_transform(data[['age_group']])
```

```python
#this is a combination of sex and age_group, there’s no inherent order between different combinations, so one-hot encoding is better:
data = pd.get_dummies(data, columns=['sex_age_group'], drop_first=True)
```

```python
# identifying unique values in column for encoding slope_recoded
print(data['slope_recoded'].unique())
```

```python
data['slope_recoded'] = data['slope_recoded'].replace(0, 'unknown')
```

# Training and Testing Models

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log_model = LogisticRegression(max_iter=10000)  # Set max_iter here
log_model.fit(x_train, y_train)
y_pred_log = log_model.predict(x_test)

print(f"Logistic Regression Accuracy: {accuracy_score(y_test, y_pred_log)}")
```

```python
#Random Forest (Higher accuracy, feature importance)
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
y_pred_rf = rf_model.predict(x_test)

print(f"Random Forest Accuracy: {accuracy_score(y_test, y_pred_rf)}")
```

```python
#Support Vector Machine (SVM)
from sklearn.svm import SVC

svm_model = SVC(kernel='rbf')
svm_model.fit(x_train, y_train)
y_pred_svm = svm_model.predict(x_test)

print(f"SVM Accuracy: {accuracy_score(y_test, y_pred_svm)}")
```

```python
#Evaluating the Models
from sklearn.metrics import classification_report

print("Logistic Regression:\n", classification_report(y_test, y_pred_log))
print("Random Forest:\n", classification_report(y_test, y_pred_rf))
print("SVM:\n", classification_report(y_test, y_pred_svm))
```

```python
# 1. First ensure your Random Forest model is trained
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train_scaled, y_train)

# 2. Define the prediction function (ONCE)
def predict_heart_disease(model, scaler):
    print("Enter the following details:")
    
    # Collect all 13 features
    age = float(input("Age: "))
    sex = int(input("Sex (0 = Female, 1 = Male): "))
    cp = int(input("Chest Pain Type (0-3): "))
    trestbps = float(input("Resting Blood Pressure: "))
    chol = float(input("Cholesterol Level: "))
    fbs = int(input("Fasting Blood Sugar > 120 mg/dl (0 = No, 1 = Yes): "))
    restecg = int(input("Resting ECG Results (0-2): "))
    thalach = float(input("Max Heart Rate Achieved: "))
    exang = int(input("Exercise Induced Angina (0 = No, 1 = Yes): "))
    oldpeak = float(input("ST depression induced by exercise: "))
    slope = int(input("Slope of peak exercise ST segment (0-2): "))
    ca = int(input("Number of major vessels (0-3): "))
    thal = int(input("Thalassemia (1-3): "))

    # Create PROPER numpy array with correct syntax
    input_data = np.array([[
        age, sex, cp, trestbps, chol, fbs, restecg, 
        thalach, exang, oldpeak, slope, ca, thal
    ]])
    
    # Scale features
    scaled_data = scaler.transform(input_data)
    
    # Make prediction
    prediction = model.predict(scaled_data)
    
    # Output result
    if prediction[0] == 1:
        print("~ High risk of heart disease! Please consult a doctor.")
    else:
        print("[] Low risk of heart disease. Keep up a healthy lifestyle!")

predict_heart_disease(rf_model, scaler)  
```
